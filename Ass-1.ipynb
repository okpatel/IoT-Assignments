```python
# Loading the dataset
df = pd.read_csv('Data.csv', encoding='utf-8', engine='python')
df.head()
```

**Question 1: How many records have a date that is expressed without using
alphabets?**


```python
# Defining the regex pattern for purely numeric dates (for example, 12-05-2023, 2023/01/01, etc.)
regex_date_pattern = r'\b\d{1,4}[-/.]\d{1,2}[-/.]\d{1,4}\b'
```

```python
# Apply the regex to each row and create a boolean column indicating if a numeric date is found
df['has_num_date'] = df['Data'].apply(lambda x: bool(re.search(regex_date_pattern, str(x))))
```

```python
# Count how many rows contain a numeric date
count_numeric_dates = df['has_num_date'].sum()
```

```python
# Display the result
print("Number of records with dates expressed without using alphabets:", count_numeric_dates)
```


**Question 2: Records with a word starting with the letter “w”**

```python
# Define the regex pattern for words starting with 'w' or 'W'
wpattern = r'\b[wW]\w*'
```

```python
# Apply the regex to each row and create a boolean column indicating if a match is found
df['has_word_starting_with_w'] = df['Data'].apply(lambda x: bool(re.search(wpattern, str(x))))
```

```python
# Count how many rows contain such a word
words_starting_with_w = df['has_word_starting_with_w'].sum()
```

```python
# Display the result
print("Number of records with a word starting with the letter 'w':", words_starting_with_w)
```
